```python
import pandas as pd

df1 = pd.DataFrame({"A":[1,2], "B":[3,4], "C":[5,6]})
df2 = pd.DataFrame({"B":[7,8], "C":[9,10], "D":[11,12]})

print(df1)
print('====')
print(df2)
```

```python
common_cols = df1.columns.intersection(df2.columns)
# 取出共同欄位
print(df1[common_cols])
print(df2[common_cols])
```

```python
df1 = pd.DataFrame({"A":[1,2,3]}, index=["app1","app2","app3"])
df2 = pd.DataFrame({"B":[4,5,6]}, index=["app2","app3","app4"])

print(df1)
print('=============')
print(df2)
```

```python
common_idx = df1.index.intersection(df2.index)

# 取出共同 index 的 row
print(df1.loc[common_idx])
print('=============')
print(df2.loc[common_idx])
```

```python
s1 = pd.Series([1,2,3], index=["x","y","z"])
s2 = pd.Series([4,5,6], index=["y","z","w"])
print(s1)
print('=============')
print(s2)

```

```python
print(s1.index.intersection(s2.index))
```
